```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Quick and dirty plot, not recommended though
plt.figure()
plt.plot(np.random.randn(50).cumsum(), 'k--')
plt.show()
```

```python
# Improved version of former
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
# PLot command, just y-axis without explicitly definig x-axis
ax1.plot(np.random.randn(50).cumsum(), 'k--')
fig
```

```python
# Figure with multilpe axes, things explicitly defined

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.plot(np.arange(50), np.random.randn(50).cumsum(), 'k--')
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
ax3.plot(np.arange(50), np.random.randn(50).cumsum(), 'k--',
         np.arange(50), np.random.randn(50).cumsum(), 'g-')
ax4.hist(np.random.randn(100),bins=20, color = 'r',alpha=0.3)

fig
```
